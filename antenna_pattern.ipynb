```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits import mplot3d
```

```python
theta = np.linspace(0,np.pi,1000)
phi = np.linspace(-np.pi,np.pi,1000)
thetav, phiv = np.meshgrid(theta, phi)
```

```python
Fp=0.5*(1+np.cos(thetav)**2)*np.cos(2*phiv)
Fc=np.cos(thetav)*np.sin(2*phiv)
```

```python
s = np.sqrt(Fp**2+Fc**2)
```

```python
x=s*(np.sin(thetav)*np.cos(phiv))
y=s*(np.sin(thetav)*np.sin(phiv))
z = s*np.cos(thetav)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
ax.view_init(30, 30)
ax.plot([0,1],[0,0],[0,0],color='black',linewidth=4.0,alpha=0.5)
ax.plot([0,0],[0,1],[0,0],color='black',linewidth=4.0,alpha=0.5)
ax.plot_surface(x,y,z, cmap='jet',edgecolor='black') #'rainbow'

#ax.contour3D(x,y,z, 100, cmap='Pastel2') #'rainbow'
```
