```python
import numpy as np
import os
from collections import Counter
import pandas as pd
```

```python
gc =pd.read_csv('gc_training.txt')
nogc =pd.read_csv('no_gc_training.txt')



```

```python
both=pd.concat([nogc['nPixels'],gc['nPixels']],axis=1,keys=['Default','Generalised Clustering'])
```

```python
plot=both.plot(kind='hist',bins=100,xlim=[0,1500],ylim=[0,100])
fig = plot.get_figure()
```

```python
plot=both.plot(kind='box',ylim=[0,150],title="Number of Pixels in Training Set Triggers")
plot.set_ylabel("Number of Pixels")
fig = plot.get_figure()
```
